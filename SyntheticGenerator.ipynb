```python
real_df = pd.read_excel("Bank_Personal_Loan_Modelling.xlsx", sheet_name="Data")

real_df = real_df.sample(2000, random_state=42)

print(real_df.shape)
real_df.head()
```

```python
income_mean = real_df["Income"].mean()
income_std = real_df["Income"].std()
```

```python
n = 500
syn = pd.DataFrame()
```

```python
syn["Income"] = np.random.normal(income_mean, income_std, n)
syn["Income"] = syn["Income"].abs()
```

```python
syn["Loan_Amount"] = syn["Income"] * np.random.uniform(1.5, 3.5, n)

syn["Loan_Amount"] += np.random.normal(0, 10, n)

zero_mask = np.random.rand(n) < 0.3
syn.loc[zero_mask, "Loan_Amount"] = 0

syn["Loan_Amount"] = syn["Loan_Amount"].clip(0, 300)
```

```python
syn["Age"] = np.random.randint(21, 65, n)
syn["Experience"] = syn["Age"] - np.random.randint(18, 25, n)
```

```python
syn["Loan_to_Income"] = syn["Loan_Amount"] / (syn["Income"] + 1)
```

```python
tenure = 5
syn["EMI"] = syn["Loan_Amount"] / tenure
```

```python
syn["FOIR"] = syn["EMI"] / (syn["Income"] + 1)
syn["FOIR"] = syn["FOIR"].clip(0, 1)
```

```python
syn["Savings"] = syn["Income"] - syn["EMI"]
syn["Savings"] = syn["Savings"].clip(lower=0)
```

```python
syn["Eligible"] = np.where(
    (syn["FOIR"] <= 0.5) &
    (syn["Loan_to_Income"] < 5) &
    (syn["Savings"] > 0),
    1,
    0
)
```

```python
print(syn.head())
print(syn.describe())
```

```python
syn.to_csv("synthetic_data.csv", index=False)
print("Saved successfully")
```
